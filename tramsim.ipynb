```python
import numpy as np
import pandas as pd
import scipy.stats as st
from collections import Counter
from tqdm import tqdm
```

```python
class Person():
    def __init__(self,choice) -> None:
        self.choice = choice
    def __repr__(self):
        return f"{self.choice}"
        
def sim(average_people=10,num_stations = 10,way="ramzey"):
    passengers = []
    
    stations = [f"station_{i}" for i in range(num_stations)]

    people_entering = np.random.poisson(average_people,num_stations)
    for idx in range(len(stations)-1):
        if way=="ramzey": 
            new_people = np.random.choice(stations[idx+1:], size=people_entering[idx])

        else :
            potential_people = np.random.choice(stations, size=people_entering[idx])

            new_people = [person for person in potential_people if person>stations[idx] ]

        
        passengers.extend(new_people)
    
    num_passengers = len(passengers)
    count = Counter(passengers)
    
    for station,occ in count.items():
        count[station]=occ/num_passengers
    
    return count


def simulations(iterations=100,average_people=10,num_stations = 10,way="ramzey"):
    sims = []

    for i in range(iterations):
        sims.append(sim(average_people=average_people,num_stations=num_stations,way=way))

    data = pd.DataFrame(sims)

    data = data.fillna(0)
    # data["sum"]=data.sum(axis=1)
    return data

data = simulations(10000,average_people=50,num_stations=5)
```

```python
data2 = simulations(iterations=10000,way="imad",average_people=50,num_stations=5)
```
